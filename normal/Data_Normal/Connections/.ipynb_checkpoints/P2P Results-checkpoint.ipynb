```python
def QoS(uuid):
    with open('15Jan/'+uuid+'.json') as f:
      peer2 = json.load(f)
#     print(json.dumps(peer2, indent=4, sort_keys=True))
    print(peer2[0]['joining_timestamp'])
    start_time = peer2[0]['joining_timestamp']

    with open('15Jan/'+uuid+'_connections.json') as f:
      peer1 = json.load(f)
    for npeer in peer1:
        print(npeer)
        print(len(peer1[npeer]))
        size = len(peer1[npeer])
        tame = [0] * size
        packl = [0] * size
        packr = [0] * size
        i=0
        for obj in peer1[npeer]:
            if(obj['from']==npeer and obj['Video']['packetsReceived'] != 0):
                tame[i] = int((obj['Video']['timestamp']-start_time)/1000)
                packl[i] = int(obj['Video']['packetsLost'])
                packr[i] = int(obj['Video']['packetsReceived'])
                i=i+1
            elif(obj['Video']['packetsReceived'] == 0):
                tame = tame[:-1]
                packl = packl[:-1]
                packr = packr[:-1]
        #print(packr)
        #print(packl)
        packt = list(map(add, packr, packl))
        #print(packt)
#         plt.plot(tame, packr)
#         plt.xlabel('Time') 
#         plt.ylabel('Packet Received') 
#         plt.show()

#         plt.plot(tame, packt)
#         plt.xlabel('Time') 
#         plt.ylabel('Packet Total') 
#         plt.show()
        
        qos = list(map(truediv, packr, packt))
        plt.plot(tame, qos)
        plt.xlabel('Time') 
        plt.ylabel('QoS') 
        plt.show()
```
